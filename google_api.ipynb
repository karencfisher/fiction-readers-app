```python
from dotenv import load_dotenv
import requests
import json
import os
```

```python
url = 'https://www.googleapis.com/books/v1/volumes?'
load_dotenv()
key = os.getenv('GOOGLE_KEY')
query = {'q':'inauthor:"Virginia Woolf"intitle:"Orlando"', 'key': key}
results = requests.get(url, params=query)
results
```
